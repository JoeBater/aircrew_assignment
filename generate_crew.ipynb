```python
import pandas as pd
import numpy as np
import random

from faker import Faker
```

```python
# LHR - London, Heathrow
# CDG - Paris, Charles De Gaulle
# AMS - Amsterdam, Schiphol
# FRA - Frankfurt
# BCN - Barcelona, El Prat 
# MXP - Milan, Malpensa
# DUB - Dublin
# CPH - Copenhagen, Kastrup

# Sample airport and aircraft pools
airports = ["LHR", "CDG", "AMS", "FRA", "BCN", "MXP", "DUB", "CPH"]
aircraft_types = ["A320", "A319", "B737"]
```

```python
N = 50

fake = Faker()

captains = [
    {"id": f"C{i}", "name": fake.name(),
     "role": 'captain',
    "base": random.choice(airports), 
    "qualified": list(set(random.choices(aircraft_types, k=np.random.randint(1, len(aircraft_types)+1))))}
    for i in range(N)
]

first_officers = [
    {"id": f"FO{i}", "name": fake.name(),
     "role": "first_officer",
    "base": random.choice(airports), 
    "qualified": list(set(random.choices(aircraft_types, k=np.random.randint(1, len(aircraft_types)+1))))}
    for i in range(N)
]

first_officers
```

```python
crew = pd.DataFrame(captains+first_officers)

crew.to_json('data/crew.json')
```
